```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.resources import INLINE
from bokeh.palettes import Spectral6
```

```python
output_notebook()
```

```python
from sklearn.decomposition import PCA
from sklearn import datasets
iris = datasets.load_iris()

pca = PCA(n_components = 2)
pca.fit(iris.data)
print(pca.explained_variance_ratio_)
```

```python
output = pca.transform(iris.data)
x = output[:,0]
y = output[:,1]
```

```python
p = figure()
colors = [Spectral6[i] for i in iris.target]
p.circle(x, y, color = colors)
show(p)
```

## Toinen datasetti

```python
def get_dataset(param, target=-1, drop=1, delim=","):
    import numpy as np
    import pandas as pd
    from io import StringIO
    import requests
    response = requests.get("https://archive.ics.uci.edu/ml/machine-learning-databases/%s/%s.data" % (param, param))
    input_ = StringIO(response.text)
    data = pd.read_csv(input_, na_values=["?"], prefix="X", header=None, sep=delim)
    data = data.dropna()
    print(data.shape)
    data = data.drop(data.columns[drop], 1)
    data = pd.to_numeric(data, errors="coerce").values # remove first N, 
    y = data[:,target]  # separate the last one as it's assumed to be the target
    np.delete(data, target, 1)
    x = data #
    return x, y
```

```python
data, target = get_dataset("breast-cancer-wisconsin")
pca = PCA(n_components = 2)
pca.fit(data)
print(pca.explained_variance_ratio_)
output = pca.transform(data)
x = output[:,0]
y = output[:,1]
```

```python
p = figure()
colors = [Spectral6[i] for i in data.target]
p.circle(x, y, color = colors)
show(p)
```
