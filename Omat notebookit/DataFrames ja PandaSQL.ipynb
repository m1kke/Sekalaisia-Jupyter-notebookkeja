```python
q = "SELECT name, age FROM titanic_train WHERE age BETWEEN 60 AND 61;"
sqldf(q, globals())
```

### Ryhmittely

```python
foo_bar = pandas.DataFrame({
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "bar"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": numpy.random.randn(8),
        "D": numpy.random.randn(8)
})
foo_bar

#tehtävä: tutustu numpy.randomin eri alifunktioihin
```

```python
foo_bar.sum()
```

```python
foo_bar.groupby("A").sum()
```

```python
grouped = foo_bar.groupby( ["A", "B"] ).sum()
grouped
```
