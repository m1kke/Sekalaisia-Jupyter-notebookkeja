## Glyphs - geometria ja viivat

```python
# piirretään koordinaatisto
p = figure(plot_width = 400, plot_height = 400)

# piirretään ympyrät
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size = 20, color = "navy", alpha=0.5)
p.square([2, 4, 8], [5, 9, 3], size=30, color="red", alpha=1)
p.triangle(7,8, size=25, color="black", alpha=0.2)

#visualisoidaan kuva
show(p)
```

```python
# luodaan koordinaatisto
k = figure(plot_width = 500, plot_height = 500)

# lisätään viiva koordinaatistoon
k.line([1, 2, 3, 4, 5, 6, 7], [2, 2, 4, 6, 6, 4, 8], line_width = 3)
k.multi_line([[1,3,2], [3,4,4,6]], [[2,1,4], [4,7,8,5]], color=["green", "yellow"], alpha=[1, 1], line_width=2)

# aktivoidaan kuva
show(k)
```

## Glyphs - Alueet

```python
i = figure(plot_width = 400, plot_height = 400)

i.patch([1, 2, 3, 4, 4.6, 5], [6, 7, 6, 7, 4, 3], alpha = 0.75, line_width=2)
i.patches([[1, 3, 2, 7], [3, 4, 6, 6]], [[2, 4, 5, 6], [2, 4, 8, 2]], color=["red", "green"], alpha=[1, 0.5], line_width=2)

show(i)
```

```python
# otetaan pi käyttöön
from math import pi

# piirretään koordinaatisto perustyökaluilla
o = figure(plot_width = 500, plot_height = 500)

# sijoitetaan kuviot figureen
o.quad( top = [2, 3, 4], bottom = [1, 2, 3], left = [1, 2, 3], right = [1.2, 2.5, 3.7], color = "#B8FEF6" )
o.rect( x = [1, 2, 3], y = [1, 2, 3], width=0.2, height = 50, color = "#EBEBEB", angle = pi/3, height_units = "screen")

# aktivoidaan figure
show(o)
```

```python
#palluroita

from bokeh.charts import Dot

data = {
    'sample': ['1st', '2nd', '1st', '2nd', '1st', '2nd'],
    'interpreter': ['python', 'python', 'pypy', 'pypy', 'jpython', 'jpython'],
    'timing': [-2, 5, 12, 40, 22, 30],
}

dots = Dot(data, values="timing", label="interpreter", group = "sample", agg = "mean", title = "Otsikko", 
           legend = "top_left", width = 550)

show(dots)
```

## Chartit

### Pylväs

```python
from bokeh.charts import Bar
from bokeh.sampledata.autompg import autompg as df

#pylväsdiagrammi
y = Bar(df, 'cyl', values='mpg', title = "Hyvä otsikko")
u = Bar(df, 'yr', values='mpg', agg='mean', color="navy", bar_width=0.4, title="Keskiverto MPG")

# selvittäkää mitä aggregate tarkoittaa

show(y)
```

```python
# aggregate soveltaminen
u = Bar(df, 'yr', values='mpg', agg='mean', color="navy", bar_width=0.4, title="Keskiverto MPG")

show(u)
```

```python
# ryhmittely
e = Bar(df, 'yr', values='mpg', agg='median', group = "origin", bar_width=1, title="Median with group", legend = "top_right")

show(e)
```

```python
# pinoamista
a = Bar(df, label = 'origin', values='mpg', agg='mean', stack = "cyl", title="AVG mpg stacked", legend = "top_right")

show(a)
```

### Box Plot

```python
from bokeh.charts import BoxPlot

b = BoxPlot(df, values = "mpg", label = "cyl", color = "#3e3e3e", title = "MPG box")
c = BoxPlot(df, values = "mpg", label = "cyl", color = "cyl", outliers = False, title = "MPG box")

show(b)
show(c)
```

### Histogrammi

```python
from bokeh.charts import Histogram

t = Histogram(df['mpg'], title = "Jakauma", bins = 50)

show(t)
```

```python
v = Histogram(df, values = "hp", color = "cyl", title = "Väritesti", legend = "top_right")

show(v)
```

### Scatter

```python
import numpy as np

from bokeh.layouts import gridplot
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT

def datetime(x):
    return np.array(x, dtype = np.datetime64)

p1 = figure(x_axis_type = "datetime", title = "Stock closing price")
p1.grid.grid_line_alpha = 0.3
p1.xaxis.axis_label = "Date"
p1.yaxis.axis_label = "Price"

p1.line(datetime(AAPL['date']), AAPL['adj_close'], color = "#A6CEE3", legend = "AAPL")
p1.line(datetime(GOOG['date']), GOOG['adj_close'], color = "#B2DF8A", legend = "GOOG")
p1.line(datetime(IBM['date']), IBM['adj_close'], color = "#33A02C", legend = "IBM")
p1.line(datetime(MSFT['date']), MSFT['adj_close'], color = "#FB9A99", legend = "MSFT")
p1.legend.location = "top_left"

aapl = np.array(AAPL['adj_close'])
aapl_dates = np.array(AAPL['date'], dtype = np.datetime64)

window_size = 30
window = np.ones(window_size)/float(window_size)
aapl_avg = np.convolve(aapl, window, 'same')

p2 = figure(x_axis_type = "datetime", title = "AAPL One-Month Average")
p2.grid.grid_line_alpha = 0
p2.xaxis.axis_label = "Date"
p2.yaxis.axis_label = "Price"
p2.ygrid.band_fill_color = "olive"
p2.ygrid.band_fill_alpha = 0.1

p2.circle(aapl_dates, aapl, size=4, legend = "close", color = "darkgrey", alpha = 0.2)
p2.line(aapl_dates, aapl_avg, legend = "avg", color = "navy")
p2.legend.location = "top_left"

show(gridplot([[p1,p2]], plot_width = 400, plot_height = 400))
```
