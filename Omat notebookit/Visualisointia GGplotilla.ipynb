```python
# kirjastot
import pandas as pd
from ggplot import *
```

```python
#ladataan datasetti
budget = pd.read_csv("mn-budget-detail-2014.csv")

#sortataan data määrän perusteella
budget = budget.sort('amount', ascending = False)[:10]
```

```python
%matplotlib inline
```

```python
# luoda plot

# mysteeri, saa selvittää miksi ei toimi

p = ggplot( budget, aes(x = "detail", y = "amount")) + geom_bar(stat = "bar", labels = budget["detail"].tolist()) + \
    gtitle("MN Capital Budget") + xlab("Spending Detail") + ylab("Amount") + scale_y_continuous(labels="millions") + \
    heme(axis_text_x = element_text(angle = 90))
print(p)
```

```python
import pandas
import numpy as np

df = pandas.DataFrame( {
        "x": np.random.normal(0, 10, 1000),
        "y": np.random.normal(0, 10, 1000),
        "z": np.random.normal(0, 10, 1000)
})

df = pandas.melt(df)

ggplot(aes(x = "value", color = "variable"), data = df) + geom_histogram()
```
