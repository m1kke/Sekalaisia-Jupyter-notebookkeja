```python
import pandas as pd
```

```python
budget = pd.read_csv("mn-budget-detail-2014.csv")
budget = budget.sort("amount", ascending = False)[:10]
```

```python
pd.options.display.mpl_style = "default"

budget_plot = budget.plot(kind = "bar", x = budget["detail"], title = "Otsikko", legend = False)
```
